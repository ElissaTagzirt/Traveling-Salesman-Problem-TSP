# **Solution 1 : Utilisation de l'arbre couvrant minimal**


L'idée générale de cette solution est de sélectionner initialement un sous-graphe qui parcourt tous les sommets mais qui ne forme pas encore un cycle hamiltonien. Ensuite, nous tentons de ameliorer ce sous-graphe en éliminant progressivement les arêtes des nœuds ayant un degré supérieur à 2 tout en ajoutant une autre arête à chaque étape pour garantir un retour à la ville initiale. Cette approche permet de trouver plus rapidement un cycle hamiltonien en simplifiant progressivement le graphe initial.

On peut modéliser la solution en trois stratégie clé :
1. *Inialisation :*
L'initialisation se fait en générant une solution initiale à l'aide de l'heuristique MST (Minimum Spanning Tree) dans la méthode mst_heuristic(). Cette solution est utilisée comme point de départ pour l'algorithme.

2. *Stratégie de séparation :*
La stratégie de séparation est implémentée dans la méthode branch_and_bound() où de nouveaux chemins sont explorés en supprimant des arrêts incidentes des sommets ayant un degré >2 et en ajoutant de nouveaux sommets pour garder le chemin de retour au sommet initial et en mettant à jour le coût courant.

3. *Fonction d'Évaluation :*
La fonction d'évaluation est utilisée pour estimer le coût du chemin actuel. Cela est fait dans la boucle principale de l'algorithme dans laquelle estimated_cost est calculé en estimant le coût de la solution partielle.

***Voici un résumé des étapes de l'algorithme :***
- Initialiser le paramètre M ou bien la borne supérieure.
- Soit  x  le noeud racine.
- En utilisant l'algorithme de Prim, cherchons  A  un arbre couvrant de poids minimal. Sans le noeud racine  x . On a une  évaluation initiale .
- Sur cet arbre, nous devrons d'abord chercher deux arêtes pour compléter le cycle jusqu'au sommet racine  x .
- *#* Sur cette configuration, on va prendre l'un des sommets qui a  degré > 2 .
- On cherche à supprimer des arretes en plus mais tout en garantissant qu'on a un cycle.
- On évalue chaque suppression, nous allons par la suite élager celle qui donne une évaluation =  Meilleure évaluation actuelle . Ou bien une évaluation  =  Fonction Objectif .
- On choisira donc que les configurations avec une évaluation  Meilleure évaluation actuelle .
- On termine quand il n'y a plus de sommets de degré plus de deux et y a plus de configurations meilleures.
***Nous allons revenir à étape #***

De ce fait, la séparation ce fait selon les sommets qui ont un degré de plus de 2 car on cherche un cycle hamiltonien. Le pracours est en largeur. L'évaluation se fait par le poids de cycle.
L'élagage repose sur les sous problème qui donne une évaluation égale ou superieure à l'évaluation précédente.


# **Solution 2 : Approche récursive de l'algorithme Branch and Bound**

::Voici un résumé de chaque point que vous avez mentionné pour votre algorithme Branch and Bound :

1. **Stratégie de parcours :**
   - L'algorithme utilise une approche de parcours exhaustif à travers les différents chemins possibles dans le graphe.
   - Il explore les successeurs du nœud courant en calculant une borne inférieure pour chaque chemin, puis en les explorant en fonction de cette borne pour déterminer les chemins prometteurs.

2. **Fonction d'évaluation :**
   - La fonction d'évaluation est utilisée pour calculer une borne inférieure pour chaque chemin exploré.
   - Elle utilise les coûts des arêtes sortantes des sommets pour calculer cette borne.
   - Deux fonctions principales sont utilisées pour calculer le premier et le deuxième coût minimum des arêtes sortantes d'un sommet.
   Voir les fonctions : ` first_min(adj, i) ` et ` second_min(adj, i) `

3. **Élagage :**
   - L'élagage est réalisé en calculant une borne inférieure pour chaque chemin exploré.
   - Si la borne inférieure dépasse le meilleur résultat trouvé jusqu'à présent, la branche est abandonnée et l'algorithme n'explore pas plus en profondeur cette branche. **Si** condition d'évaluation satisfaite : Aller en profondeur. **Sinon** élager.
   - Cela permet de réduire le nombre de chemins à explorer.

4. **Initialisation :**
   - L'initialisation comprend plusieurs étapes :
     - Initialisation de la borne inférieure initiale en calculant une somme des deux plus petits coûts sortants de chaque sommet, puis en arrondissant cette somme à l'entier supérieur.
     - Marquage du premier sommet comme visité et initialisation du chemin courant avec ce sommet.
     - Appel à la fonction récursive principale pour démarrer l'exploration.

# **Solution 3 : Amélioration de la solution 2 avec la programmation dynamique**

L'amélioration de la deuxième solution du Problème du Voyageur de Commerce en utilisant la méthode Branch and Bound avec **une table de mémorisation** ` memo_table ` au niveau de la fonction ` first_min ` et `second_min ` représente une avancée significative. Cette amélioration vise à réduire la redondance des calculs en stockant les résultats intermédiaires dans une table, accessibles par une clé unique représentant l'état du problème. Lors de l'exploration de nouveaux chemins partiels, le système vérifie d'abord si le résultat pour cet état est déjà présent dans la table, évitant ainsi des calculs redondants. Cette optimisation permet d'accélérer considérablement l'algorithme, en particulier pour les problèmes de grande taille, en réduisant la complexité temporelle et en améliorant l'efficacité des bornes inférieures. Les résultats obtenus démontrent une amélioration significative des performances, avec une exécution près de 10 fois plus rapide par rapport au code original non amélioré. En somme, l'introduction de la table de mémorisation représente une étape essentielle dans l'optimisation de l'algorithme Branch and Bound pour résoudre le Problème du Voyageur de Commerce.

# **Solution 1 :**

```python
import numpy as np  # Importation de la bibliothèque numpy pour les calculs numériques
import heapq  # Importation de heapq pour utiliser une file de priorité
import time

class Graph:
    def __init__(self, matrix):
        self.V = len(matrix)  # Nombre de sommets dans le graphe
        self.graph = matrix  # Matrice d'adjacence représentant le graphe

    # Fonction heuristique pour générer une solution initiale à l'aide de l'algorithme Prim
    def mst_heuristic(self):
        mst_parent = self.primMST(0, set())  # Crée un arbre couvrant de poids minimum

        stack = [0]  # Utilisé pour le parcours en profondeur
        visited = set([0])  # Enregistre les sommets visités
        path = []  # Chemin partiel

        # Parcours en profondeur pour générer une solution initiale
        while stack:
            current_vertex = stack.pop()
            path.append(current_vertex)
            for neighbor in range(self.V):
                # Vérifie si le voisin n'a pas été visité et s'il existe un lien avec le sommet actuel
                if neighbor != current_vertex and neighbor not in visited and self.graph[current_vertex][neighbor] > 0:
                    stack.append(neighbor)
                    visited.add(neighbor)

        # Ajoute le sommet de départ pour compléter le cycle
        path.append(0)

        # Calcule le coût total de la solution initiale
        total_cost = sum(self.graph[path[i]][path[i + 1]] for i in range(len(path) - 1))

        return path, total_cost

    # Algorithme de Prim pour trouver un arbre couvrant de poids minimum
    def primMST(self, start, visited):
        key = [np.inf] * self.V  # Initialise les clés à l'infini
        parent = [-1] * self.V  # Initialise les parents à -1
        mstSet = [False] * self.V  # Initialise l'ensemble d'arêtes de l'arbre couvrant à False
        key[start] = 0  # Initialise la clé du premier sommet à 0
        pq = [(0, start)]  # File de priorité pour stocker les sommets

        # Boucle principale pour construire l'arbre couvrant
        while pq:
            _, u = heapq.heappop(pq)
            mstSet[u] = True

            for v in range(self.V):
                if self.graph[u][v] > 0 and mstSet[v] == False and key[v] > self.graph[u][v] and v not in visited:
                    # Met à jour la clé et le parent
                    key[v] = self.graph[u][v]
                    parent[v] = u
                    heapq.heappush(pq, (key[v], v))

        return parent

    # Calcule le coût d'un parcours
    def calculate_cost(self, parent):
        cost = 0
        for i in range(self.V):
            if parent[i] != -1:
                cost += self.graph[i][parent[i]]
        return cost

    # Algorithme de Branch and Bound pour résoudre le TSP
    def branch_and_bound(self):
        # Génère une solution initiale à l'aide de l'heuristique MST
        initial_solution, initial_cost = self.mst_heuristic()
        best_solution = initial_solution
        best_cost = initial_cost
        stack = [([0], 0, 0)]

        # Boucle principale de l'algorithme de Branch and Bound
        while stack:
            path, last, current_cost = stack.pop()
            if len(path) == self.V:
                # Complète le cycle
                if self.graph[last][0] > 0:  # Vérifie s'il y a un chemin de retour au sommet de départ
                    total_cost = current_cost + self.graph[last][0]
                    if total_cost < best_cost:
                        best_solution = path + [0]
                        best_cost = total_cost
            else:
                for next in range(self.V):
                    if next not in path and self.graph[last][next] > 0:
                        new_path = path + [next]
                        new_cost = current_cost + self.graph[last][next]
                        # Estime le coût restant en utilisant l'heuristique MST
                        remaining = self.primMST(next, set(new_path))
                        estimated_cost = self.calculate_cost(remaining) + new_cost
                        if estimated_cost < best_cost:
                            stack.append((new_path, next, new_cost))

        return best_solution, best_cost


graph = Graph(adj)

start_time = time.time()
solution, cost = graph.branch_and_bound()
end_time = time.time()
elapsed_time = end_time - start_time

print(f"Elapsed Time: {elapsed_time} seconds")
# Continue with the branch and bound algorithm
print("Best solution:", solution)
print("Minimum cost:", cost)
```

```python
import time  # Importation du module time pour calculer le temps d'exécution
import math  # Importation du module math pour utiliser la fonction ceil

# Fonction pour copier le chemin courant dans le chemin final et ajouter le premier sommet à la fin pour former un cycle
def copy_to_final(curr_path):
    final_path[:N + 1] = curr_path[:]  # Copie le chemin courant dans final_path
    final_path[N] = curr_path[0]  # Ajoute le premier sommet à la fin pour compléter le cycle

# Fonction pour trouver le coût minimum d'une arête sortante du sommet i
def first_min(adj, i):
    # Retourne le coût minimum parmi tous les arcs sortants du sommet i, à l'exception de l'arc vers lui-même
    return min(adj[i][j] for j in range(N) if i != j)

# Fonction pour trouver le deuxième coût minimum d'une arête sortante du sommet i
def second_min(adj, i):
    values = [adj[i][j] for j in range(N) if i != j]  # Liste des coûts des arêtes sortantes du sommet i
    values.sort()  # Trie la liste des coûts
    return values[1]  # Retourne le deuxième coût le plus bas

# Fonction récursive principale de l'algorithme Branch and Bound
def TSP_rec(adj, curr_bound, curr_weight, level, curr_path, visited):
    global final_res  # Référence à la variable globale pour le résultat final
    # Si tous les sommets ont été visités, vérifie si le chemin forme un cycle valide
    if level == N:
        if adj[curr_path[level - 1]][curr_path[0]] != 0:  # Vérifie s'il y a un retour au point de départ
            curr_res = curr_weight + adj[curr_path[level - 1]][curr_path[0]]  # Calcule le coût total du cycle
            if curr_res < final_res:  # Si le coût total est inférieur au meilleur coût trouvé jusqu'à présent
                copy_to_final(curr_path)  # Copie le chemin courant dans le chemin final
                final_res = curr_res  # Met à jour le meilleur coût
        return

    # Boucle pour explorer les successeurs du nœud courant
    for i in range(N):
        if adj[curr_path[level - 1]][i] != 0 and not visited[i]:  # Si le sommet i est accessible et non visité
            temp = curr_bound  # Sauvegarde la borne courante
            curr_weight += adj[curr_path[level - 1]][i]  # Ajoute le coût de l'arête au poids courant

            # Ajuste la borne pour le niveau suivant
            if level == 1:
                curr_bound -= (first_min(adj, curr_path[level - 1]) + first_min(adj, i)) / 2
            else:
                curr_bound -= (second_min(adj, curr_path[level - 1]) + first_min(adj, i)) / 2

            # Si la borne inférieure actuelle + le poids courant est inférieure au meilleur résultat
            if curr_bound + curr_weight < final_res:
                curr_path[level] = i  # Ajoute le sommet i au chemin courant
                visited[i] = True  # Marque le sommet i comme visité

                # Appel récursif pour explorer le niveau suivant
                TSP_rec(adj, curr_bound, curr_weight, level + 1, curr_path, visited)

            # Réinitialise le poids courant et la borne pour explorer d'autres branches
            curr_weight -= adj[curr_path[level - 1]][i]
            curr_bound = temp

            # Réinitialise le tableau des sommets visités pour le prochain appel
            visited = [False] * len(visited)
            for j in range(level):
                if curr_path[j] != -1:
                    visited[curr_path[j]] = True

# Fonction pour initialiser et exécuter l'algorithme TSP
def TSP(adj):
    curr_bound = 0  # Borne initiale
    curr_path = [-1] * (N + 1)  # Chemin courant initialisé à -1
    visited = [False] * N  # Tableau des sommets visités initialisé à False

    # Calcule la borne initiale
    for i in range(N):
        curr_bound += (first_min(adj, i) + second_min(adj, i))
    curr_bound = math.ceil(curr_bound / 2)  # Arrondit la borne à l'entier supérieur

    visited[0] = True  # Marque le premier sommet comme visité
    curr_path[0] = 0  # Commence le chemin par le premier sommet

    # Appelle la fonction récursive pour commencer l'exploration
    TSP_rec(adj, curr_bound, 0, 1, curr_path, visited)

# Code principal
if __name__ == "__main__":

    adj =[
    [0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    [10, 0, 25, 35, 15, 20, 30, 5, 40, 35, 45, 50, 55, 60],
    [15, 25, 0, 5, 10, 30, 40, 15, 25, 30, 20, 45, 30, 35],
    [20, 35, 5, 0, 40, 25, 15, 10, 20, 45, 30, 35, 40, 50],
    [25, 15, 10, 40, 0, 20, 10, 25, 30, 35, 50, 55, 60, 65],
    [30, 20, 30, 25, 20, 0, 35, 30, 5, 15, 40, 45, 50, 55],
    [35, 30, 40, 15, 10, 35, 0, 20, 25, 30, 45, 60, 65, 70],
    [40, 5, 15, 10, 25, 30, 20, 0, 35, 40, 55, 35, 40, 45],
    [45, 40, 25, 20, 30, 5, 25, 35, 0, 15, 10, 25, 30, 35],
    [50, 35, 30, 45, 35, 15, 30, 40, 15, 0, 25, 20, 15, 20],
    [55, 45, 20, 30, 50, 40, 45, 55, 10, 25, 0, 15, 30, 35],
    [60, 50, 45, 35, 55, 45, 60, 35, 25, 20, 15, 0, 10, 15],
    [65, 55, 30, 40, 60, 50, 65, 40, 30, 15, 30, 10, 0, 25],
    [70, 60, 35, 50, 65, 55, 70, 45, 35, 20, 35, 15, 25, 0]
    ]
    N = len(adj)  # Nombre de sommets dans le graphe
    final_path = [None] * (N + 1)  # Chemin final
    visited = [False] * N  # Sommets visités
    final_res = float('inf')  # Meilleur résultat initialisé à l'infini

    start_time = time.time()  # Début du chronométrage
    TSP(adj)  # Exécution de l'algorithme TSP
    end_time = time.time()  # Fin du chronométrage

    # Affichage des résultats
    print(f"Minimum cost: {final_res}")
    print(f"Path Taken: {final_path}")
    print(f"Time taken: {end_time - start_time} seconds")
```

```python
import time
import math

# Initialisation de la valeur infinie utilisée pour représenter un coût très élevé, inatteignable en pratique.
maxsize = float('inf')

# N représente le nombre de villes dans le problème du TSP.
N = len(adj)
# final_path va stocker le chemin final (le cycle complet) qui a le coût minimum.
final_path = [None] * (N + 1)
# visited est un tableau de booléens pour vérifier si une ville a été visitée ou non.
visited = [False] * N
# final_res stocke le coût minimum du chemin trouvé jusqu'à présent.
final_res = maxsize

# memo_table est utilisée pour la mémorisation, afin d'éviter les recalculs des minimums pour chaque ville.
memo_table = {}

# Cette fonction copie le chemin courant dans le chemin final une fois qu'un chemin complet est trouvé.
def copyToFinal(curr_path):
    final_path[:N + 1] = curr_path[:]
    final_path[N] = curr_path[0]  # Pour faire de cela un cycle, le dernier ville renvoie à la première.

# Trouve la valeur minimale (coût) pour aller de la ville i à n'importe quelle autre ville.
def firstMin(adj, i):
    # Si cette valeur a déjà été calculée, la retourner directement pour éviter un recalcul.
    min_val = memo_table.get((i, 1), None)
    if min_val is not None:
        return min_val

    # Sinon, calculer le minimum et le stocker dans memo_table.
    min_val = min(val for idx, val in enumerate(adj[i]) if idx != i)
    memo_table[(i, 1)] = min_val
    return min_val

# Trouve la deuxième plus petite valeur (coût) pour aller de la ville i à n'importe quelle autre ville.
def secondMin(adj, i):
    # Utilisation de memo_table pour le caching.
    min_val = memo_table.get((i, 2), None)
    if min_val is not None:
        return min_val

    # Initialisation des deux plus petits coûts à l'infini.
    first, second = maxsize, maxsize
    for j, val in enumerate(adj[i]):
        if i != j:
            if val <= first:
                second = first
                first = val
            elif val <= second and val != first:
                second = val

    # Mise à jour de memo_table avec la deuxième plus petite valeur.
    memo_table[(i, 2)] = second
    return second

# Fonction récursive pour le parcours en profondeur avec branch and bound.
def TSPRec(adj, curr_bound, curr_weight, level, curr_path, visited):
    global final_res

    # Si tous les nœuds sont visités, vérifier si il y a un retour au point de départ pour compléter le cycle.
    if level == N:
        # Vérifier si le dernier nœud peut retourner au premier pour former un cycle.
        if adj[curr_path[level - 1]][curr_path[0]] != 0:
            curr_res = curr_weight + adj[curr_path[level - 1]][curr_path[0]]
            if curr_res < final_res:
                copyToFinal(curr_path)
                final_res = curr_res
        return

    # Essayer de visiter chaque ville non visitée et explorer le chemin.
    for i in range(N):
        if adj[curr_path[level-1]][i] != 0 and not visited[i]:
            temp = curr_bound
            curr_weight += adj[curr_path[level - 1]][i]

            # Ajustement de la limite inférieure pour le prochain nœud.
            if level == 1:
                curr_bound -= ((firstMin(adj, curr_path[level - 1]) + firstMin(adj, i)) / 2)
            else:
                curr_bound -= ((secondMin(adj, curr_path[level - 1]) + firstMin(adj, i)) / 2)

            # Continuer à explorer ce chemin uniquement si le coût estimé reste inférieur au meilleur coût trouvé jusqu'à présent.
            if curr_bound + curr_weight < final_res:
                curr_path[level] = i
                visited[i] = True
                TSPRec(adj, curr_bound, curr_weight, level + 1, curr_path, visited)

            # Réinitialiser le poids et la borne pour l'exploration du prochain nœud.
            curr_weight -= adj[curr_path[level - 1]][i]
            curr_bound = temp
            # Réinitialiser le tableau visited.
            visited = [False] * N
            for j in range(level):
                if curr_path[j] != -1:
                    visited[curr_path[j]] = True

# La fonction principale qui configure et lance l'algorithme TSP.
def TSP(adj):
    # Initialiser la limite inférieure pour le chemin partant de la première ville.
    curr_bound = 0
    curr_path = [-1] * (N + 1)
    visited = [False] * N

    # Calculer la limite inférieure initiale pour l'ensemble du chemin.
    for i in range(N):
        curr_bound += (firstMin(adj, i) + secondMin(adj, i))
    curr_bound = math.ceil(curr_bound / 2)

    visited[0] = True
    curr_path[0] = 0

    # Commencer le TSP à partir de la première ville.
    TSPRec(adj, curr_bound, 0, 1, curr_path, visited)


# Démarrer le chronomètre pour mesurer le temps d'exécution de l'algorithme.
start_time = time.time()
TSP(adj)
end_time = time.time()
elapsed_time = end_time - start_time

# Affichage du temps d'exécution, du coût minimum trouvé et du chemin correspondant.
print(f"Elapsed Time: {elapsed_time} seconds")
print("Minimum cost:", final_res)
print("Path Taken:", final_path)
```

```python
import networkx as nx
import numpy as np
import time
from scipy.spatial import distance_matrix
from networkx.algorithms.matching import min_weight_matching
from networkx.algorithms.tree.mst import minimum_spanning_tree
import heapq
from xml.etree import ElementTree as ET



def christofides_algorithm(G):
    # Trouver un arbre couvrant minimal
    T = nx.minimum_spanning_tree(G, weight='weight')

    # Trouver les sommets de degré impair dans l'arbre couvrant minimal
    odd_degree_nodes = [node for node in T.nodes() if T.degree(node) % 2 == 1]

    # Restreindre le graphe aux sommets de degré impair
    odd_graph = G.subgraph(odd_degree_nodes)

    # Trouver un matching parfait de poids minimum dans le graphe restreint
    matching = min_weight_matching(odd_graph)

    # Ajouter le matching à l'arbre couvrant minimal pour obtenir un multigraphe
    multigraph = nx.MultiGraph(T)
    for edge in matching:
        multigraph.add_edge(edge[0], edge[1], weight=G[edge[0]][edge[1]]['weight'])

    # Trouver un circuit Eulerien dans le multigraphe
    eulerian_circuit = list(nx.eulerian_circuit(multigraph))

    # Convertir le circuit Eulerien en un circuit Hamiltonien et calculer le coût
    visited = set()
    hamiltonian_circuit = []
    total_cost = 0
    last_node = None
    #print(eulerian_circuit)
    for u, v in eulerian_circuit:
        if u not in visited:
            if last_node is not None:
                total_cost += G[last_node][u]['weight']
            visited.add(u)
            hamiltonian_circuit.append(u)
            last_node = u
        if v not in visited:
            if last_node is not None:
                total_cost += G[last_node][v]['weight']
            visited.add(v)
            hamiltonian_circuit.append(v)
            last_node = v
    # Ajouter le coût pour retourner au point de départ
    # Ajouter le coût pour retourner au point de départ pour former un cycle
    if hamiltonian_circuit:
        start_node = hamiltonian_circuit[0]
        end_node = hamiltonian_circuit[-1]
        if start_node != end_node:
            hamiltonian_circuit.append(start_node)
            total_cost += G[end_node][start_node]['weight']

    return hamiltonian_circuit, total_cost



# Exemple d'utilisation :
# G doit être défini comme un graphe pondéré NetworkX
# start_node est le nœud de départ
start_time = time.time()
start_node = 0
circuit, total_cost = christofides_algorithm(G)
end_time = time.time()

execution_time = end_time - start_time
print("Circuit Hamiltonien :", circuit)
print("Coût total :", total_cost)
print("Temps d'exécution :", execution_time, "secondes")
```
